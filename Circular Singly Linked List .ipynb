```python


class Node:
    def __init__(self,val):
        self.val = val
        self.next = None

class CSLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0


# printing linked list value using str
    def __str__(self):
        temp_node = self.head
        result = ""
        while temp_node :
            result += str(temp_node.val)
            temp_node = temp_node.next
            if temp_node ==  self.head:
                break
            result += '->'
        return result

    
# append in circular linked list
    def append(self,val):
        new_node = Node(val)
        if self.length == 0:
            self.head= new_node
            self.tail = new_node
            new_node.next = new_node 
        else:
            self.tail.next = new_node
            new_node.next = self.head
            self.tail =new_node
        self.length += 1



# Prepend in CSLL
    def prepend(self,val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            self.tail=new_node
            new_node.next = self.head
        else:
            new_node.next = self.head
            self.head = new_node
            self.tail.next = new_node
        self.length +=1


# insert in CSLL
    def insert(self,val,index):
        new_node=Node(val)
        if index < 0 or index >self.length:
            return False
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
            new_node.next  = new_node
        elif index == 0:
            new_node.next = self.head
            self.head = new_node
            self.tail.next = new_node
        elif index == self.length:
            self.tail.next = new_node
            new_node.next = self.head
            self.tail = new_node
        else:
            temp_node = self.head
            for _ in range(index-1):
                temp_node = temp_node.next
            new_node.next = temp_node.next
            temp_node.next = new_node
        self.length += 1
        return True




    def traverse(self):
        curr_node = self.head
        while curr_node:
            print(curr_node.val)
            curr_node = curr_node.next
            if curr_node == self.head:
                break

    def search(self,target):
        current = self.head
        while current:
            if current.val == target:
                return True
            current = current.next
            if current == self.head:
                break
        return False

    def get(self,index):
        if index<-1 or index >= self.length:
            return None
        elif index == -1:
            return self.tail
        current = self.head
        for _ in range(index):
            current = current.next
        return current

    def set_value(self,value,index):
        temp = self.get(index)
        if temp:
            temp.val = value
            return True
        return False


    def pop_first(self):
        if self.length  == 0:
            return False
        popped_node = self.head
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.tail.next = self.head
            popped_node.next = None
        self.length -= 1
        return popped_node
        
    def pop(self):
        popped_node = self.tail
        temp = self.head
        while temp.next is not self.tail:
            temp = temp.next
        self.tail = temp
        temp.next = self.head
        popped_node.next = None
        self.length -=1
        return popped_node

    def remove(self,index):
        if index == 0:
            return self.pop_first()
        if index <1 or index >=self.length:
            return None
        if index == self.length-1 or index == -1:
            return self.pop()
        prev_node = self.get(index-1)
        popped_node = prev_node.next
        prev_node.next = popped_node.next
        popped_node.next = None
        self.length -=1


    def delete_all(self):
        if self.length == 0:
            return
        self.tail.next = None
        self.head = None
        self.tail = None
        self.length = 0
            
        
cslinkedlist = CSLinkedList()
cslinkedlist.append(10)
cslinkedlist.append(20)
cslinkedlist.append(30)
cslinkedlist.append(50)
cslinkedlist.prepend(22)
cslinkedlist.insert(222,3)
# print(cslinkedlist)
# print(cslinkedlist.head.val)
# print(cslinkedlist.tail.val)
# print(cslinkedlist.traverse())
# print(cslinkedlist.head.next.val)
# print(cslinkedlist.search(50))
print(cslinkedlist)
print(cslinkedlist.delete_all())
print(cslinkedlist)
cslinkedlist.delete_all()
```
